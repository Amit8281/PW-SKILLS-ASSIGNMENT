## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python library for statistical data visualization. Here are five commonly used plots in Seaborn along with their typical use cases:

- **Bar Plot**: A bar plot represents categorical data using rectangular bars, where the length of each bar corresponds to the value of the category. It is useful for comparing the values of different categories.
Use cases: Comparing categorical variables, showing counts or frequencies.

- **Histogram**: A histogram displays the distribution of a continuous variable by dividing it into bins and showing the frequency or count of data points within each bin. It helps visualize the shape and spread of the data.
Use cases: Visualizing data distribution, identifying outliers, understanding data ranges.

- **Scatter Plot**: A scatter plot displays the relationship between two continuous variables by plotting individual data points on a 2D plane. It helps identify patterns, correlations, or clusters in the data.
Use cases: Exploring relationships between variables, identifying trends, detecting outliers.

- **Box Plot**: A box plot (or box-and-whisker plot) summarizes the distribution of a continuous variable by displaying the quartiles, median, and outliers. It provides a concise visual representation of the data distribution and highlights potential outliers.
Use cases: Comparing distributions, detecting outliers, visualizing summary statistics.

- **Heatmap**: A heatmap represents data in a tabular format using colors to indicate the values. It is commonly used to display correlation matrices or visualize the relationship between two categorical variables.
Use cases: Exploring relationships between variables, visualizing matrices, highlighting patterns or clusters.

These are just a few examples of the plots available in Seaborn. The library provides many more plot types and customization options to cater to various visualization needs.



## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset from seaborn
fmri = sns.load_dataset("fmri")

# Create a line plot using Seaborn and Matplotlib
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Display the plot
plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
### Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset from seaborn
titanic = sns.load_dataset("titanic")

# Create the first box plot for 'age'
plt.subplot(1, 2, 1)  # Create a subplot for the first box plot
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age by Passenger Class')

# Create the second box plot for 'fare'
plt.subplot(1, 2, 2)  # Create a subplot for the second box plot
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare by Passenger Class')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset from seaborn
diamonds = sns.load_dataset("diamonds")

# Create a histogram with hue using seaborn
sns.histplot(data=diamonds, x='price', hue='cut', element='step', kde=True)

# Set labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices by Cut')

# Show the plot
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns

# Load the iris dataset from seaborn
iris = sns.load_dataset("iris")

# Create a pair plot with hue using seaborn
sns.pairplot(data=iris, hue="species")

# Display the plot
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset from seaborn
flights = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Create a heatmap using seaborn
sns.heatmap(flights_pivot, cmap="YlGnBu")

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Flights Heatmap")

# Show the plot
plt.show()
```
